```python
import pandas as pd
import pickle
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
preprocess_model = pickle.load(open('preprocess_model.pkl','rb'))
```

```python
x_train = preprocess_model[0]
y_train = preprocess_model[1]
x_test = preprocess_model[2]
y_test = preprocess_model[3]
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(x_train,y_train)
```

```python
print(regression.coef_)
print(regression.intercept_)
```

```python
# prediction for test data
reg_pred= regression.predict(x_test)
```

```python
reg_pred
```

```python
# R2 score or R-squared score
from sklearn.metrics import r2_score
linear_r2 = r2_score(y_test,reg_pred)
print(linear_r2)
```

```python
# adjusted R square score
adjusted_linear_r2 = 1 - (1-linear_r2)*(len(y_test)-1)/(len(y_test)-x_test.shape[1]-1)
adjusted_linear_r2
```

## Ridge regression

```python
from sklearn.linear_model import Ridge
ridge = Ridge()
ridge.fit(x_train,y_train)
```

```python
ridge_pred = ridge.predict(x_test)
```

```python
ridge_pred
```

```python
# Let's see the R2 score for Ridge
from sklearn.metrics import r2_score
ridge_r2 = r2_score(y_test,ridge_pred)
print(ridge_r2)
```

```python
# let's see the Adjusted R2 score
adjusted_ridge_r2 = 1 - (1-ridge_r2)*(len(y_test)-1)/(len(y_test)-x_test.shape[1]-1)
adjusted_ridge_r2 
```

## Lasso regression

```python
from sklearn.linear_model import Lasso
lasso = Lasso()
lasso.fit(x_train,y_train)
```

```python
lasso_pred = lasso.predict(x_test)
```

```python
# R2 score for Lasso
from sklearn.metrics import r2_score
lasso_r2 = r2_score(lasso_pred,y_test)
lasso_r2
```

```python
# Adjusted R2 score for Lasso
adjusted_lasso_r2 = 1 - (1-lasso_r2)*(len(y_test)-1)/(len(y_test)-x_test.shape[1]-1)
adjusted_lasso_r2
```

## Now talk about assumption of linear regression
linear relation between input and output feature.

No multicolinearity.

residuals should be Normally distributed.

Homoscedasticity - spread of residual in graph should be equal.

No autocorrelation of errors - no autocorrelation in plot of residuals.

```python
plt.figure(figsize=(15,15))
for i in range(x_test.shape[1]):
    plt.subplot(3,3,i+1)
    plt.xlabel('x_train')
    plt.ylabel('y_train')
    plt.scatter(x_train.T[i],y_train)
```

2. no multicollinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = []

for i in range(x_train.shape[1]):
    vif.append(variance_inflation_factor(x_train,i))
```
